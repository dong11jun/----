```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

class Transaction:
    def __init__(self, date, category, description, amount, transaction_type):
        self.date = date
        self.category = category
        self.description = description
        self.amount = amount
        self.transaction_type = transaction_type

class Account:
    def __init__(self, name):
        self.name = name
        self.balance = 0
        self.transactions = []

    def deposit(self, date, amount, category, description):
        transaction_date = datetime.strptime(date, "%Y-%m-%d")
        self.balance += amount
        transaction = Transaction(transaction_date, category, description, amount, "입금")
        self.transactions.append(transaction)
        print(f"{self.name}님, {amount}원이 {category} 카테고리로 {transaction_date.date()}에 입금되었습니다.")

    def withdraw(self, date, amount, category, description):
        transaction_date = datetime.strptime(date, "%Y-%m-%d")
        if self.balance >= amount:
            self.balance -= amount
            transaction = Transaction(transaction_date, category, description, amount, "출금")
            self.transactions.append(transaction)
            print(f"{self.name}님, {amount}원이 {category} 카테고리에서 {transaction_date.date()}에 출금되었습니다.")
        else:
            print("잔액이 부족합니다.")

    def show_balance(self):
        print(f"{self.name}님의 현재 잔액은 {self.balance}원입니다.")
        print("최근 거래 내역:")
        recent_transactions = self.transactions[-5:]  # 최근 5개 거래만 선택
        for transaction in recent_transactions:
            print(transaction.date, transaction.category, transaction.description, transaction.amount)

    def save_transactions_to_csv(self):
        filename = f"{self.name}_transactions.csv"
        with open(filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Date", "Type", "Category", "Description", "Amount"])
            for transaction in self.transactions:
                writer.writerow([transaction.date.date(), transaction.transaction_type, transaction.category, transaction.description, transaction.amount])
        print(f"{filename}에 거래 내역이 저장되었습니다.")

        self.visualize_transaction_history()

    def visualize_transaction_history(self):
        dates = [transaction.date for transaction in self.transactions]
        amounts = [transaction.amount for transaction in self.transactions]
        df = pd.DataFrame({"Date": dates, "Amount": amounts})

        plt.plot(df["Date"], df["Amount"])
        plt.xlabel("Date")
        plt.ylabel("Amount")
        plt.title(f"{self.name}님의 거래 내역")
        plt.xticks(rotation=45)
        plt.show()


class FamilyAccount(Account):
    def __init__(self, name, family_name):
        super().__init__(name)
        self.family_name = family_name

    def load_transactions_from_csv(self):
        filename = f"{self.name}_transactions.csv"
        try:
            df = pd.read_csv(filename)
            for _, row in df.iterrows():
                date = datetime.strptime(row["Date"], "%Y-%m-%d")
                category = row["Category"]
                description = row["Description"]
                amount = row["Amount"]
                transaction_type = row["Type"]
                transaction = Transaction(date, category, description, amount, transaction_type)
                self.transactions.append(transaction)
                if transaction_type == "입금":
                    self.balance += amount
                elif transaction_type == "출금":
                    self.balance -= amount
        except FileNotFoundError:
            print(f"{filename}을 찾을 수 없습니다.")

        self.show_balance()


family_members = [
  FamilyAccount("아빠", "김창우"),
  FamilyAccount("엄마", "김문주"),
  FamilyAccount("누나", "김효정"),
  FamilyAccount("나", "김동준"),
  FamilyAccount("동생", "김동규")
]

for member in family_members:
    member.load_transactions_from_csv()

while True:
  print("가족 구성원 목록:")
  for i, member in enumerate(family_members):
    print(f"{i + 1}. {member.name}")

  member_choice = input("거래를 진행할 구성원 번호를 선택하세요 (종료: 0): ")
  member_choice = int(member_choice)

  if member_choice == 0 or member_choice > len(family_members):
    print("프로그램을 종료합니다.")
    break

  member = family_members[member_choice - 1]

  print(f"선택한 구성원: {member.name}")
  print("1. 입금")
  print("2. 출금")
  print("3. 가계부 확인")
  print("4. 거래 내역 저장 및 시각화")

  action_choice = input("실행할 작업 번호를 선택하세요: ")
  action_choice = int(action_choice)

  if action_choice == 1:
    date = input("날짜를 입력하세요 (예: 2023-05-23): ")
    amount = int(input("입금할 금액을 입력하세요: "))
    category = input("카테고리를 입력하세요: ")
    description = input("설명을 입력하세요: ")
    member.deposit(date, amount, category, description)

  elif action_choice == 2:
    date = input("날짜를 입력하세요 (예: 2023-05-23): ")
    amount = int(input("출금할 금액을 입력하세요: "))
    category = input("카테고리를 입력하세요: ")
    description = input("설명을 입력하세요: ")
    member.withdraw(date, amount, category, description)

  elif action_choice == 3:
    member.show_balance()

  elif action_choice == 4:
    member.save_transactions_to_csv()

  else:
    print("잘못된 작업 번호입니다. 다시 선택해주세요.")
#2023142008 김동준 010-3876-6889
```
